```python
import networkx as nx
import pandas as pd
import matplotlib as plot

%matplotlib inline
```

```python
dependency = pd.read_csv("./dependency/Test-Table 1.csv", names=["item","dependency"])
```

```python
G = nx.DiGraph()

for row in dependency.itertuples():
    index, item, required_item = row
    G.add_edge(item, required_item)
```

```python
T = nx.dfs_tree(G,'Science pack 3')
```

```python
pos = nx.spring_layout(T,iterations=1000)
nx.draw(T,labels=True,pos=pos)
```
